```python
def read_csv_safely(path, **kwargs):
    try:
        return pd.read_csv(path, **kwargs)  # UTF-8 (default)
    except UnicodeError:  # broader & correct
        return pd.read_csv(path, encoding="cp1252", **kwargs)


def load_tashlomim_and_hazmanot(path, type):
    df = read_csv_safely(path)  

    # Clean column names FIRST
    df.columns = (df.columns
                    .str.strip()
                    .str.lower()
                    .str.replace('-', '_')
                    .str.replace(' ', '_'))

    # Select expected schema and remove weird columns
    if type == "tashlomim":
        df.columns = EXPECTED_TASHLOMIM_COLUMNS

    elif type == "hazmanot":
        df.columns = EXPECTED_HAZMANOT_COLUMNS
        df = df[df["fund_code"] != "#"]

    else:
        raise ValueError(f"Unknown type '{type}'")

    # Apply dtypes
    if type == "tashlomim":
        df = df.astype({
            "order_id": "string",
            "po_2": "Int64",
            "coin": "Int64",
            "budget_year": "Int64"
        }, errors="raise")

    if type == "hazmanot":
        df = df.astype({
            "order_id": "string",
            "fund_code": "string",
            "order_starting_date": "Int64"
        }, errors="raise")

    return df
tashlomim_data_orginal_from_user = load_tashlomim_and_hazmanot(tashlomim_path, "tashlomim")
hazmanot_data_original_from_user = load_tashlomim_and_hazmanot(hazmanot_path, "hazmanot")
```

# check what fc4 there is under each zroa and what fc3 unfer each huka

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

MIN_MIMUN = 1_000_000  # minimum mimun to show

# =====================================================
# Prepare data once
# =====================================================
df = (
    tashlomim_data_specific_coin
    .groupby(["zroa", "fc4", "huka", "fc3"], as_index=False)
    .agg({"mimun": "sum"})
)

# =====================================================
# Shared plotter class
# =====================================================
class FeaturePlotter:
    def __init__(self, df):
        self.df = df

    def plot(self, mode, parent_value):
        clear_output(wait=True)
        display(mode_dropdown, parent_dropdown)

        if mode == "fc4 in zroa":
            parent_col = "zroa"
            child_col = "fc4"
            title = f"Mimun by FC4 | Zroa = {parent_value}"
        else:  # fc3 in huka
            parent_col = "huka"
            child_col = "fc3"
            title = f"Mimun by FC3 | Huka = {parent_value}"

        # group and filter only positive, >= MIN_MIMUN
        plot_df = (
            self.df[self.df[parent_col] == parent_value]
            .groupby(child_col)["mimun"]
            .sum()
        )

        plot_df = plot_df[plot_df >= MIN_MIMUN].sort_values(ascending=False)

        if plot_df.empty:
            print(f"No {child_col.upper()} with mimun ≥ {MIN_MIMUN/1e6:.0f}M for {parent_value}")
            return

        ax = plot_df.plot(kind="bar", figsize=(12, 4), title=title)
        ax.set_xlabel(child_col.upper())
        ax.set_ylabel("Mimun")

        for i, v in enumerate(plot_df):
            ax.text(i, v, f"{v/1e6:.1f}M", ha="center", va="bottom", fontsize=9)

        plt.tight_layout()
        plt.show()


# =====================================================
# Widgets
# =====================================================
mode_dropdown = widgets.Dropdown(
    options=["fc4 in zroa", "fc3 in huka"],
    description="Mode:",
    style={"description_width": "initial"}
)

parent_dropdown = widgets.Dropdown(
    description="Value:",
    style={"description_width": "initial"}
)

plotter = FeaturePlotter(df)


# =====================================================
# Widget logic
# =====================================================
def update_parent_options(*args):
    """Update the dropdown with only parents that have total mimun >= MIN_MIMUN"""
    if mode_dropdown.value == "fc4 in zroa":
        # sum mimun per zroa
        sums = df.groupby("zroa")["mimun"].sum()
        parent_dropdown.options = sorted(sums[sums >= MIN_MIMUN].index.tolist())
    else:
        sums = df.groupby("huka")["mimun"].sum()
        parent_dropdown.options = sorted(sums[sums >= MIN_MIMUN].index.tolist())

    # select first value
    if parent_dropdown.options:
        parent_dropdown.value = parent_dropdown.options[0]
    else:
        parent_dropdown.value = None

def update_plot(change=None):
    if parent_dropdown.value is not None:
        plotter.plot(mode_dropdown.value, parent_dropdown.value)
    else:
        clear_output(wait=True)
        display(mode_dropdown, parent_dropdown)
        print(f"No parent with mimun ≥ {MIN_MIMUN/1e6:.0f}M to plot.")

# Attach observers
mode_dropdown.observe(update_parent_options, names="value")
parent_dropdown.observe(update_plot, names="value")

# =====================================================
# Initial render
# =====================================================
update_parent_options()
display(mode_dropdown, parent_dropdown)
update_plot()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ----------------------------------------
# Subjects
# ----------------------------------------
subjects_hazmanot = [
    "sales_office",
    "sales_organistion",
    "sales_document_type",
    "sales_group",
    "inviting_partner"
]

subjects_tashlomim = ["huka", "zroa", "fc4", "fc3"]
all_subjects = subjects_hazmanot + subjects_tashlomim

year_options = ["All", 2021, 2022, 2023, 2024, 2025]

# ----------------------------------------
# Widgets
# ----------------------------------------
year_dropdown = widgets.Dropdown(
    options=year_options,
    value=2025,
    description="Year:",
    style={"description_width": "initial"}
)

subject_dropdown = widgets.Dropdown(
    options=all_subjects,
    value=all_subjects[0],
    description="Subject:",
    style={"description_width": "initial"}
)

topn_slider = widgets.IntSlider(
    value=5,
    min=1,
    max=20,
    step=1,
    description="Top N:",
    style={"description_width": "initial"}
)

# ----------------------------------------
# Helper functions
# ----------------------------------------
def filter_year(df, year, year_col="budget_year"):
    if year == "All":
        return df[df[year_col].isin([2021, 2022, 2023, 2024, 2025])]
    return df[df[year_col] == year]


def add_percent_labels(ax, values):
    total = values.sum()
    ymax = values.max()

    for i, val in enumerate(values):
        pct = val / total * 100
        ax.text(
            i,
            val + ymax * 0.02,   # always just above the bar
            f"{pct:.1f}%\n({val/1e6:.2f}M)",
            ha="center",
            va="bottom",
            fontsize=10
        )

# ----------------------------------------
# Main plot function
# ----------------------------------------
def plot_subject(year, subject, top_n):
    clear_output(wait=True)
    display(year_dropdown, subject_dropdown, topn_slider)

    # Dataset & value column
    if subject in subjects_hazmanot:
        df = filter_year(hazmanot_data_specific_coin, year)
        value_col = "sales_order_sum"
        title_base = "Sales Order Sum"
    else:
        df = filter_year(tashlomim_data_specific_coin, year)
        value_col = "mimun"
        title_base = "Mimun"

    if df.empty:
        print("No data for selected year.")
        return

    # Group by subject
    group_data = (
        df.groupby(subject, dropna=True)[value_col]
        .sum()
        .reset_index()
    )

    group_data[value_col] = pd.to_numeric(group_data[value_col], errors="coerce")
    group_data = group_data[group_data[value_col] > 0]

    if group_data.empty:
        print("No positive values to display.")
        return

    # Top N
    top_data = (
        group_data
        .sort_values(value_col, ascending=False)
        .head(top_n)
        .sort_values(value_col, ascending=False)
    )

    # Explicit order (CRITICAL FIX)
    order = top_data[subject].tolist()

    # Dynamic figure width
    fig_width = min(1 + 1.5 * len(top_data), 16)
    plt.figure(figsize=(fig_width, 6))

    ax = sns.barplot(
        data=top_data,
        x=subject,
        y=value_col,
        order=order,
        palette=sns.color_palette("husl", len(top_data))
    )

    plt.title(
        f"Top {top_n} {subject.replace('_',' ').title()} by {title_base} ({year})",
        fontsize=16,
        fontweight="bold"
    )

    plt.xlabel(subject.replace("_", " ").title(), fontsize=12)
    plt.ylabel(title_base, fontsize=12)

    add_percent_labels(ax, top_data[value_col].values)

    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

# ----------------------------------------
# Widget interactions
# ----------------------------------------
year_dropdown.observe(
    lambda change: plot_subject(year_dropdown.value, subject_dropdown.value, topn_slider.value),
    names="value"
)

subject_dropdown.observe(
    lambda change: plot_subject(year_dropdown.value, subject_dropdown.value, topn_slider.value),
    names="value"
)

topn_slider.observe(
    lambda change: plot_subject(year_dropdown.value, subject_dropdown.value, topn_slider.value),
    names="value"
)

# ----------------------------------------
# Initial display
# ----------------------------------------
display(year_dropdown, subject_dropdown, topn_slider)
plot_subject(year_dropdown.value, subject_dropdown.value, topn_slider.value)
```

```python
class HukaDetailPlotter:
    def __init__(self, tashlomim_df, active_hukas_list):
        """
        tashlomim_df: original dataframe with columns
            ['payment_year','huka','expediture_type','zroa','doc_type','mimun']
        active_hukas_list: list of active hukas (from HukaPlotter)
        """
        self.tashlomim_df = tashlomim_df
        self.active_hukas_list = active_hukas_list
        self.huka_df = None
        self.fc_or_zroa_pivot = None
        self.doc_type_pivot = None
        self.expenditure_type_specific_huka = None
    
    def prepare_huka_data(self, huka_to_plot, fc_or_zroa_selected):
        """
        Prepares all the aggregated data for a specific huka.
        
        fc_or_zroa_selected: str, column name to use instead of 'zroa'
        """
        # Filter for active hukas and group
        active_huka_data = self.tashlomim_df[
            self.tashlomim_df["huka"].isin(self.active_hukas_list)
        ].groupby(["payment_year", "huka", "expediture_type", fc_or_zroa_selected, "doc_type"]) \
         .agg({"mimun": "sum"}).reset_index()

        # Filter for the specific huka
        self.huka_df = active_huka_data[active_huka_data["huka"] == huka_to_plot].sort_values("payment_year")

        # Aggregate by dynamic fc_or_zroa_selected and doc_type
        fc_or_zroa_data = self.huka_df.groupby(["payment_year", fc_or_zroa_selected]).agg({"mimun": "sum"}).reset_index()
        doc_type_data = self.huka_df.groupby(["payment_year", "doc_type"]).agg({"mimun": "sum"}).reset_index()
        self.expenditure_type_specific_huka = self.huka_df.groupby(["expediture_type"]).agg({"mimun": "sum"}).reset_index()

        # Percentage for expenditure type
        total_mimun = self.expenditure_type_specific_huka["mimun"].sum()
        self.expenditure_type_specific_huka["pracentage"] = (self.expenditure_type_specific_huka["mimun"] / total_mimun) * 100

        # Pivot using dynamic fc_or_zroa_selected
        self.fc_or_zroa_pivot = fc_or_zroa_data.pivot_table(
            index="payment_year",
            columns=fc_or_zroa_selected,
            values="mimun",
            aggfunc="sum",
            fill_value=0
        ).sort_index()

        self.doc_type_pivot = doc_type_data.pivot_table(
            index="payment_year",
            columns="doc_type",
            values="mimun",
            aggfunc="sum",
            fill_value=0
        ).sort_index()

    def plot(self):
        """
        Plots dynamic fc_or_zroa_selected stacked bar, Doc type stacked bar, and Expenditure type pie chart.
        """
        if self.huka_df is None:
            raise ValueError("Call prepare_huka_data(huka_to_plot, fc_or_zroa_selected) first!")

        # Colors
        fc_or_zroa_colors = sns.color_palette("Set2", n_colors=len(self.fc_or_zroa_pivot.columns))
        doc_type_colors = sns.color_palette("Set3", n_colors=len(self.doc_type_pivot.columns))
        expenditure_colors = sns.color_palette("Pastel1", n_colors=len(self.expenditure_type_specific_huka))

        # Explode largest pie slice
        explode = [0.1 if i == self.expenditure_type_specific_huka["pracentage"].idxmax() else 0 
                   for i in range(len(self.expenditure_type_specific_huka))]

        # Create subplots
        fig, axes = plt.subplots(1, 3, figsize=(18, 6))

        # Dynamic fc_or_zroa_selected stacked bar
        self.fc_or_zroa_pivot.plot(
            kind="bar",
            stacked=True,
            ax=axes[0],
            color=fc_or_zroa_colors,
            edgecolor="black",
            width=0.8
        )
        axes[0].set_title(f"{self.fc_or_zroa_pivot.columns.name} stacked by year", fontsize=14)
        axes[0].set_xlabel("Year")
        axes[0].set_ylabel("Total mimun")
        axes[0].grid(axis="y", linestyle="--", alpha=0.5)
        axes[0].legend(title=self.fc_or_zroa_pivot.columns.name, bbox_to_anchor=(1.05, 1), loc="upper left")

        # Doc type stacked bar
        self.doc_type_pivot.plot(
            kind="bar",
            stacked=True,
            ax=axes[1],
            color=doc_type_colors,
            edgecolor="black",
            width=0.8
        )
        axes[1].set_title("Doc type stacked by year", fontsize=14)
        axes[1].set_xlabel("Year")
        axes[1].grid(axis="y", linestyle="--", alpha=0.5)
        axes[1].legend(title="Doc type", bbox_to_anchor=(1.05, 1), loc="upper left")

        # Expenditure type pie
        self.expenditure_type_specific_huka.plot(
            kind="pie",
            y="pracentage",
            labels=self.expenditure_type_specific_huka["expediture_type"],
            autopct="%.1f%%",
            explode=explode,
            colors=expenditure_colors,
            ax=axes[2],
            legend=False,
            startangle=140,
            shadow=True
        )
        axes[2].set_title("Expenditure type (% of total)", fontsize=14)
        axes[2].set_ylabel("")

        plt.tight_layout()
        plt.show()
# Instantiate
detail_plotter = HukaDetailPlotter(tashlomim_data_specific_coin, plotter.active_hukas_list)

# Prepare data for a specific huka
huka_options = plotter.active_hukas_list
huka_plot_dropdown = widgets.Dropdown(
    options=huka_options,
    value=huka_options[0],
    description='Select huka:',
    style={'description_width': 'initial'}
)


widgets.interact(
    lambda huka: [detail_plotter.prepare_huka_data(huka, fc_or_zroa_selected), detail_plotter.plot()],
    huka=huka_plot_dropdown
)

```

```python
def choose_decomposition(ts, r2_score, elasticity):
    use_multiplicative = (r2_score > R2_THRESHOLD_FOR_REGRESSION) and (elasticity > ELASTICITY_THRESHOLD_FOR_REGRESSION)
    has_negative = (ts < 0).any()
    has_zero = (ts == 0).any()

    if has_negative:
        return "additive", "negative values present in the data"
    elif has_zero and use_multiplicative:
        return "multiplicative", "zero values present in the data"
    elif use_multiplicative:
        return "multiplicative", "variance scales with level and no zero values in the data"
    else:
        return "additive", "variance roughly constant or weakly related"

def make_decompose(ts, decomposition_type="additive", period=12, robust=True):
    """
    Decompose a time series using STL.
    
    decomposition_type:
        - "additive": STL on original series
        - "multiplicative": log1p transform + STL (recommended)
    """

    if decomposition_type == "multiplicative":
        ts_used = np.log1p(ts)
    else:
        ts_used = ts.copy()

    stl = STL(ts_used, period=period, robust=robust)
    res = stl.fit()

    return res

def extract_decomposition(decomposition):
    return decomposition.trend, decomposition.seasonal, decomposition.resid
 
def plot_decomposition(decomposition):
    fig = decomposition.plot()
    fig.set_size_inches(12, 8)  
    plt.show()
```

```python
# ==========================
# Configuration
# ==========================
fc_or_zroa_selected = 'fc3'  # replace with your dynamic column
KEYS = ['huka', 'expediture_type', fc_or_zroa_selected, 'doc_type']
FULL_KEYS = KEYS + ['paymant_date']

original_data_daily = (
    tashlomim_data_specific_coin
    .groupby(FULL_KEYS, as_index=False)
    .agg(mimun=('mimun', 'sum'))
    .sort_values('paymant_date')
)
# ==========================
# Spike Detection Class
# ==========================
class SpikeDetection:
    """
    Detects spikes in time series data, interpolates them, and allows plotting.
    """
    def __init__(self, data, keys=KEYS, rolling_window=ROLLING_WINDOW,
                 min_points=MIN_POINTS, deviation_threshold=DEVIATION_THRESHOLD):
        self.data = data.copy()
        self.keys = keys
        self.rolling_window = rolling_window
        self.min_points = min_points
        self.deviation_threshold = deviation_threshold
        self.flagged_as_spikes = None
        self.cleaned_data = None

    def interpolate_group(self, group):
        """
        Interpolates NaN values in 'mimun' using time-based interpolation.
        """
        group = group.sort_values('paymant_date').set_index('paymant_date')
        group['mimun'] = group['mimun'].interpolate(method='time')
        return group.reset_index()

    def find_spikes(self):
        """
        Detect spikes using rolling mean/std and z-score method.
        """
        df = self.data

        # Rolling mean and std
        df['rolling_mean'] = df.groupby(self.keys)['mimun'].transform(
            lambda x: x.rolling(window=self.rolling_window, min_periods=self.min_points).mean()
        )
        df['rolling_std'] = df.groupby(self.keys)['mimun'].transform(
            lambda x: x.rolling(window=self.rolling_window, min_periods=self.min_points).std()
        )

        # Deviation (z-score like)
        df['deviation'] = (df['mimun'] - df['rolling_mean']).abs() / df['rolling_std']

        # Flag spikes
        self.flagged_as_spikes = df[df['deviation'] > self.deviation_threshold]
        df['is_spike'] = df.index.isin(self.flagged_as_spikes.index)

        # Replace spikes with NaN
        df.loc[df['is_spike'], 'mimun'] = np.nan

        self.data = df
        # Interpolate cleaned data
        self.cleaned_data = df.groupby(self.keys, group_keys=False).apply(self.interpolate_group)

        return self.data, self.flagged_as_spikes, self.cleaned_data

    def summary_spikes(self):
        """
        Print a summary of spikes detected.
        """
        num_spikes = self.data['is_spike'].sum()
        percent_spikes = num_spikes / len(self.data) * 100
        print(f"Total of {num_spikes} out of {len(self.data)} rows were flagged as spikes ({percent_spikes:.2f}%).")

    def plot_sample_timeseries(self, sample_key, original_data):
        """
        Plot original vs cleaned time series and highlight spikes.
        """
        # Filter helper
        def filter_df(df):
            mask = np.ones(len(df), dtype=bool)
            for k in self.keys:
                mask &= df[k] == sample_key[k]
            return df[mask].sort_values('paymant_date')

        original = filter_df(original_data)
        cleaned = filter_df(self.cleaned_data)
        spikes = filter_df(self.flagged_as_spikes)

        plt.figure(figsize=(14, 6))
        plt.plot(original['paymant_date'], original['mimun'], label='Original', marker='o')
        plt.plot(cleaned['paymant_date'], cleaned['mimun'], label='Cleaned', marker='o')
        plt.scatter(spikes['paymant_date'], spikes['mimun'], color='red', label='Flagged Spikes', zorder=5)
        plt.title('Original vs Cleaned Mimun Time Series')
        plt.xlabel('Payment Date')
        plt.ylabel('Mimun')
        plt.legend()
        plt.tight_layout()
        plt.show()

```

```python
class TimeSeriesAnalysis:
    def __init__(self, ts):
        self.ts = ts
        self.count = ts.count()
        self.mean = ts.mean()
        self.std_before = ts.std()
        self.zeros_pct = (ts == 0).mean()
        self.cv_before = self.std_before / self.mean if self.mean != 0 else 0
        self.std_after = None
        self.cv_after = None
        self.std_improvement = None

    def _plot_summary(self, title, summary_text, warnings):
        fig, ax = plt.subplots(figsize=(7,4))
        ax.axis('off')
        ax.text(
            0.5, 0.5,
            f"{title}\n\n{summary_text}\n\n{warnings}",
            ha='center', va='center',
            fontsize=13,
            fontfamily='monospace',
            bbox=dict(boxstyle="round,pad=0.5", facecolor="#f0f0f0", edgecolor="#cccccc")
        )
        plt.show()

    def plot_ts_summary(self):
        summary_text = (
            f"Count of observations: {self.count}\n"
            f"Percentage of zeros: {self.zeros_pct:.2%}\n"
            f"Std / Mean (CV): {self.cv_before:.2%}"
        )

        warning1 = (
            "Warning: CV > 100%, data highly variable"
            if self.cv_before > 1 else
            "CV under 100%, data moderately variable"
        )

        warning2 = (
            "Warning: more than 50% zeros, sparse series"
            if self.zeros_pct > 0.5 else
            "Most data is not zero, series not sparse"
        )

        self._plot_summary("Time Series Summary", summary_text, f"{warning1}\n{warning2}")

    def plot_ts_summary_after_dealing_spikes(self, cleaned_ts):
        self.std_after = cleaned_ts.std()
        self.cv_after = self.std_after / cleaned_ts.mean() if cleaned_ts.mean() != 0 else 0
        # Compute improvement
        self.std_improvement = (self.std_before - self.std_after) / self.std_before

        summary_text = (
            f"Std / Mean (CV) after cleaning: {self.cv_after:.2%}\n"
            f"Std improvement: {self.std_improvement:.2%}"
        )

        self._plot_summary("Time Series Summary After Dealing with Spikes", summary_text, "")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.holtwinters import ExponentialSmoothing

def forecast_trend(trend_comp, periods=12, method='polynomial', poly_order=2):
    """
    Forecast STL trend component using either polynomial regression or Holt's method.
    
    Parameters
    ----------
    trend_comp : pd.Series
        STL trend component, indexed by datetime
    periods : int
        Number of future periods to forecast
    method : str
        'polynomial' or 'holt'
    poly_order : int
        Polynomial order if using polynomial regression (default 2)
    
    Returns
    -------
    pd.Series : forecasted trend values, indexed by datetime
    """
    trend_series = trend_comp.dropna().astype(float)
    n = len(trend_series)
    
    # Create future datetime index
    last_date = trend_series.index[-1]
    freq = trend_series.index.inferred_freq
    if freq is None:
        freq = pd.infer_freq(trend_series.index)
    future_index = pd.date_range(start=last_date, periods=periods+1, freq=freq)[1:]
    
    if method == 'polynomial':
        # Time as numbers
        t = np.arange(n).reshape(-1,1)
        X = np.hstack([t**i for i in range(1, poly_order+1)])
        y = trend_series.values
        
        model = LinearRegression().fit(X, y)
        
        # Forecast future
        t_future = np.arange(n, n+periods).reshape(-1,1)
        X_future = np.hstack([t_future**i for i in range(1, poly_order+1)])
        forecast = model.predict(X_future)
        
    elif method == 'holt':
        model = ExponentialSmoothing(trend_series, trend='add', seasonal=None)
        fit = model.fit()
        forecast = fit.forecast(periods)
        
    else:
        raise ValueError("Method must be 'polynomial' or 'holt'")
    
    return pd.Series(forecast, index=future_index)



```

# damped trend
###### ses is like ARIMA(0,1,1) just level
###### holt is like ARIMA(0,2,2) just level with trend 
###### holt-damped is like ARIMA(0,2,2) with trend and damping

```python
def forecast_with_damped_trend(trend_comp,shrink_strength, periods=12):

    series = trend_comp.dropna().astype(float)

    if len(series) < 3:
        raise ValueError("Not enough data to fit damped trend model")

    freq = series.index.inferred_freq
    if freq is None:
        freq = pd.infer_freq(series.index)

    model = ExponentialSmoothing(
        series,
        trend="add", 
        damped_trend=True, 
        seasonal=None
    )

    fit = model.fit(optimized=True) # find alpha

    # Extract last level and trend
    level = fit.level[-1]
    trend = fit.trend[-1]
    # ---- Dynamic shrink of negative trend ----
    # shrink the trend if it is negetive so the start of the forcast wont be so negetive
    if trend < 0:
        shrink_factor = np.exp(
            -shrink_strength * abs(trend) / (abs(level) + 1e-6)
        )
        trend = trend * shrink_factor

    # how much of the last trend is added to the level at each step
    phi = fit.params['damping_trend'] # each future trend is multiplied by this

    forecasts = []
    for h in range(1, periods + 1):
        damped_multiplier = (1 - phi**h) / (1 - phi) # Total “weight” of all future trend contributions
        forecast = level + trend * damped_multiplier
        forecasts.append(forecast)

    last_date = series.index[-1]
    future_index = pd.date_range(
        start=last_date,
        periods=periods + 1,
        freq=freq
    )[1:]

    return pd.Series(forecasts, index=future_index)

def get_non_negative_forecast(
    trend_comp,
    periods=12,
    last_n_months=3,
    initial_shrink=1.5,
    max_shrink=10,
):
    """
    Generate forecast and increase shrink_strength until
    the sum of the last `last_n_months` is non-negative.
    """

    shrink_strength = initial_shrink

    while shrink_strength < max_shrink:
        monthly_forecast = forecast_with_damped_trend(
            trend_comp,
            shrink_strength,
            periods=periods
        )

        sum_last_n = monthly_forecast[-last_n_months:].sum()

        if sum_last_n >= 0:
            break

        shrink_strength += 1

    # if still sum of total negetive make the whole series zero
    if monthly_forecast.sum() < 0:
        monthly_forecast = pd.Series(0, index=monthly_forecast.index)

    return monthly_forecast
```

```python
class TimeSeriesInspector:
    ROLLING_WINDOW = 24
    ACTIVITY_THRESHOLD = 0
    ACTIVITY_RATIO_THRESHOLD = 0.5
    QUANTILE = 0.25

    def __init__(self, ts: pd.Series):
        self.ts = ts.copy()
        self.ts.index = self.ensure_datetime_index(self.ts.index)
        self.ts.index = self.align_to_month_end(self.ts)
        self.activity_ratio = None
        self.rolling_mean = None
        self.rolling_std = None
        self.cv = None
        self.magnitude_thresh = None

    @staticmethod
    def ensure_datetime_index(index):
        if not isinstance(index, pd.DatetimeIndex):
            return pd.to_datetime(index)
        return index

    @staticmethod
    def align_to_month_end(ts: pd.Series) -> pd.DatetimeIndex:
        return ts.index + pd.offsets.MonthEnd(0)

    def compute_activity_ratio(self):
        is_active = self.ts > self.ACTIVITY_THRESHOLD
        self.activity_ratio = is_active.rolling(self.ROLLING_WINDOW, min_periods=1).mean()

    def compute_statistics(self):
        ts_positive = self.ts[self.ts > 0]
        self.magnitude_thresh = ts_positive.quantile(self.QUANTILE) if not ts_positive.empty else 0
        self.rolling_mean = self.ts.rolling(self.ROLLING_WINDOW).mean()
        self.rolling_std = self.ts.rolling(self.ROLLING_WINDOW).std()
        self.cv = self.rolling_std / self.rolling_mean.abs()

    def find_last_inactive_period(self, months=None):
        if months is None:
            months = self.ROLLING_WINDOW
        if self.activity_ratio is None:
            return None
        for date in reversed(self.activity_ratio.index):
            window = self.activity_ratio.loc[date:date + pd.DateOffset(months=months)]
            if window.values.sum() == 0:
                return window
        return None

    def determine_first_valid_date(self):
        inactive_period = self.find_last_inactive_period()
        first_date_of_active_ts = inactive_period.index[-1] if inactive_period is not None else self.ts.index[0]

        valid = (
            (self.activity_ratio >= self.ACTIVITY_RATIO_THRESHOLD) &
            (self.rolling_mean >= self.magnitude_thresh)
        )

        first_date_of_valid_ts = self.ts.index[0]
        if valid.any():
            first_date_of_valid_ts = valid[valid.values].index[0]

        return max(first_date_of_active_ts, first_date_of_valid_ts)

    def get_clean_active_ts(self):
        self.compute_activity_ratio()
        self.compute_statistics()
        first_date_to_use = self.determine_first_valid_date()
        return self.ts.loc[first_date_to_use:]
```

```python
# =====================================================
# Residual Share Model (Single Feature, Fixed)
# =====================================================
class ResidualShareModel:
    def __init__(
        self,
        tashlomim_df: pd.DataFrame,
        hazmanot_df: pd.DataFrame,
        resid_comp: pd.Series,
        clean_active_ts: pd.Series,
        feature: str,
        lags: tuple = (1, 3, 6, 12),
        model_params: dict = None,
        verbose: bool = False,
    ):
        self.tashlomim_df = tashlomim_df
        self.hazmanot_df = hazmanot_df
        self.resid_comp = resid_comp
        self.clean_active_ts = clean_active_ts
        self.feature = feature
        self.lags = lags
        self.verbose = verbose

        self.model = xgb.XGBRegressor(
            **(model_params or dict(n_estimators=500, max_depth=5, learning_rate=0.05))
        )

        self.training_df = None
        self.feature_columns = None
        self.target_col = None

    # -------------------------------------------------
    def _prepare_combined_resid(self) -> pd.DataFrame:
        combined = pd.concat(
            [self.resid_comp.rename("resid"),
             self.clean_active_ts.rename("mimun_this_month")],
            axis=1,
        ).dropna()
        return combined

    # -------------------------------------------------
    def _prepare_feature_ts(
        self,
        huka: str,
        expediture_type: str,
        fc_or_zroa_selected: str,
        zroa_or_fc3: str,
        doc_type: str,
    ) -> pd.DataFrame:
        df = self.tashlomim_df[self.tashlomim_df["order_id"] != "#"].copy()
        haz = self.hazmanot_df.groupby("order_id")[self.feature].first().reset_index()
        df = df.merge(haz, on="order_id", how="left")
        df = df[df[self.feature].notna()]
        df = df[
            (df["huka"] == huka)
            & (df["expediture_type"] == expediture_type)
            & (df[fc_or_zroa_selected] == zroa_or_fc3)
            & (df["doc_type"] == doc_type)
        ]
        ts = (
            df.groupby([pd.Grouper(key="paymant_date", freq="M"), self.feature])["mimun"]
            .sum()
            .reset_index()
            .set_index("paymant_date")
        )
        ts.rename(columns={"mimun": f"{self.feature}_mimun"}, inplace=True)
        return ts

    # ----------------------------------------------------
    def _build_training_table(self, feature_ts: pd.DataFrame, combined_resid: pd.DataFrame) -> pd.DataFrame:
        df = feature_ts.merge(combined_resid, left_index=True, right_index=True, how="left")
        df["mimun_this_month"] = df["mimun_this_month"].replace(0, np.nan)

        df[f"{self.feature}_share"] = df[f"{self.feature}_mimun"] / df["mimun_this_month"]
        df[f"{self.feature}_resid_share"] = df["resid"] * df[f"{self.feature}_share"]

        # Lags
        for lag in self.lags:
            df[f"{self.feature}_lag_{lag}"] = df.groupby(self.feature)[f"{self.feature}_resid_share"].shift(lag)

        # Categorical code
        df[f"{self.feature}_code"] = df[self.feature].astype("category").cat.codes

        # Time features
        df["month"] = df.index.month
        df["year"] = df.index.year
        df = df.drop(columns=[self.feature])

        df = df.replace([np.inf, -np.inf], np.nan)

        return df.dropna()

    # -------------------------------------------------
    def fit(
        self,
        huka: str,
        expediture_type: str,
        fc3: str,
        doc_type: str,
    ):
        combined_resid = self._prepare_combined_resid()
        feature_ts = self._prepare_feature_ts(huka, expediture_type, "fc3",fc3, doc_type)
        self.training_df = self._build_training_table(feature_ts, combined_resid)

        if self.training_df.empty:
            return "failed to build training table mot enough data"

        self.target_col = f"{self.feature}_resid_share"
        lag_cols = [f"{self.feature}_lag_{l}" for l in self.lags]
        code_cols = [f"{self.feature}_code"]
        self.feature_columns = lag_cols + ["month", "year"] + code_cols

        X = self.training_df[self.feature_columns].apply(pd.to_numeric, errors="coerce")
        y = pd.to_numeric(self.training_df[self.target_col], errors="coerce")

        valid_idx = X.notna().all(axis=1) & y.notna()
        X = X.loc[valid_idx].reset_index(drop=True)
        y = y.loc[valid_idx].reset_index(drop=True)

        #  guard against too few rows
        if len(X) < 10:
            return "failed to build training table mot enough data"

        #  guard against constant target
        if y.nunique() <= 1:
            return "target constant"

        #  guard against inf
        if np.isinf(X.values).any():
            return "inf in X"
        
        if self.training_df[f"{self.feature}_code"].nunique() <= 1:
            return "feature constant"


        self.model.fit(X, y)
        '''
        if self.verbose:
            print(f"Model for '{self.feature}' trained!")
            print("Features used:", self.feature_columns)

        print("------")
        print("Feature:", self.feature)
        print("Rows:", len(self.training_df))
        print("Unique categories:",
            self.training_df[f"{self.feature}_code"].nunique())
        print("Target unique:",
            self.training_df[self.target_col].nunique())
        print("Target std:",
            self.training_df[self.target_col].std())
        '''


        return self


class ResidualForecastAggregator:
    def __init__(
        self,
        models: dict,  # dictionary {feature_name: model}
        features: list,
        feature_columns: dict,  # {feature_name: feature_columns}
        lags: tuple = (1, 3, 6, 12),
        verbose: bool = False,
    ):
        self.models = models
        self.features = features
        self.feature_columns = feature_columns
        self.lags = lags
        self.verbose = verbose

    def predict_future(self, training_df_dict: dict, start_date="2026-01-31", periods=12) -> dict:
        """
        training_df_dict: {feature_name: training_df} 
        Returns: {feature_name: predicted Series}
        """
        future_index = pd.date_range(start=start_date, periods=periods, freq="M")
        predictions = {}

        for f in self.features:
            model = self.models[f]
            feature_cols = self.feature_columns[f]
            last_row = training_df_dict[f].iloc[-1].copy()
            future_preds = []

            for dt in future_index:
                X_row = last_row[feature_cols].values.reshape(1, -1)
                pred = model.predict(X_row)[0]
                future_preds.append(pred)

                # Update lags for next step
                for lag in reversed(self.lags):
                    col = f"{f}_lag_{lag}"
                    if lag == 1:
                        last_row[col] = pred
                    else:
                        last_row[col] = last_row[f"{f}_lag_{lag-1}"]

                # Update month/year for feature columns
                if "month" in feature_cols:
                    last_row["month"] = dt.month
                if "year" in feature_cols:
                    last_row["year"] = dt.year

            predictions[f] = pd.Series(future_preds, index=future_index, name=f)
        return predictions

    def aggregate_total(self, predictions: dict) -> pd.Series:
        """
        Sum all features’ predicted residuals month by month
        """
        total_resid = pd.concat(predictions.values(), axis=1).sum(axis=1)
        total_resid.name = "total_resid"
        return total_resid

def train_models(specific_features, lags, huka_to_inspect, expenditure_type_to_inspect, zroa_or_fc3_to_inspect, doc_type_to_inspect, resid_comp, clean_active_ts):
    models = {}
    feature_cols = {}
    training_dfs = {}
# Train models for each feature
    for feature in specific_features:
        model_obj = ResidualShareModel(
            tashlomim_data_specific_coin,
            hazmanot_data_specific_coin,
            resid_comp,
            clean_active_ts,
            feature,
            lags,
            verbose=True,
        )
        
        flag = model_obj.fit(
            huka_to_inspect,
            expenditure_type_to_inspect,
            zroa_or_fc3_to_inspect,
            doc_type_to_inspect,
        )

        if isinstance(flag, str):
            #if model_obj.verbose:
                #print(f"Skipping feature '{feature}' → {flag}")
            continue


        models[feature] = model_obj.model
        feature_cols[feature] = model_obj.feature_columns
        training_dfs[feature] = model_obj.training_df

    return models, feature_cols, training_dfs

# make it a function to train models for each feature


specific_features = ["inviting_partner", "sales_group", "sales_office", "sales_organistion"]
#specific_features = ["inviting_partner"]

lags = (1, 3, 6, 12)

# Dictionaries to hold models, feature columns, and training data
models = {}
feature_cols = {}
training_dfs = {}

models, feature_cols, training_dfs = train_models(specific_features, lags, huka_to_inspect, expenditure_type_to_inspect, zroa_or_fc3_to_inspect, doc_type_to_inspect, resid_comp, clean_active_ts)
```

```python
# =====================================================
# Safe Residual Forecast Aggregator (Silent)
# =====================================================
class ResidualForecastAggregator:
    def __init__(
        self,
        models: dict,             # dictionary {feature_name: trained model}
        features: list,           # all features we want to predict
        feature_columns: dict,    # {feature_name: feature_columns}
        lags: tuple = (1, 3, 6, 12),
        clip_std: float = 1    # how many std to clip extreme predictions
    ):
        self.models = models
        self.features = features
        self.feature_columns = feature_columns
        self.lags = lags
        self.clip_std = clip_std
        self.predictions = {}      # per-feature predictions
        self.total_resid = None    # total residuals across features

    # -------------------------------------------------
    @staticmethod
    def safe_predict(model, X_row, resid_series, last_pred=None, clip_std=1):
        """
        Predict a residual safely.
        """
        pred = model.predict(X_row)[0]

        # Clip extreme predictions to historical residual bounds
        resid_std = resid_series.std()
        resid_mean = resid_series.mean()
        lower = resid_mean - clip_std * resid_std
        upper = resid_mean + clip_std * resid_std
        pred_clipped = np.clip(pred, lower, upper)

        # Optional dampening: limit sudden jumps
        if last_pred is not None:
            max_change = 2 * resid_std
            pred_clipped = np.clip(pred_clipped, last_pred - max_change, last_pred + max_change)

        return float(pred_clipped)

    # -------------------------------------------------
    def predict_future(self, training_df_dict: dict, resid_series: pd.Series, start_date="2026-01-31", periods=12) -> dict:
        """
        Predict future residual shares for all features.
        Missing features without a trained model are filled with zeros.
        """
        future_index = pd.date_range(start=start_date, periods=periods, freq="M")
        self.predictions = {}

        for f in self.features:
            if f not in self.models or f not in training_df_dict:
                # Missing feature -> fill with zeros
                self.predictions[f] = pd.Series(0, index=future_index, name=f)
                continue

            model = self.models[f]
            cols = self.feature_columns[f]
            last_row = training_df_dict[f].iloc[-1].copy()
            preds = []

            for dt in future_index:
                X_row = last_row[cols].values.reshape(1, -1)
                last_pred_val = last_row[f"{f}_lag_1"] if f"{f}_lag_1" in last_row else None

                pred = self.safe_predict(model, X_row, resid_series, last_pred=last_pred_val, clip_std=self.clip_std)
                preds.append(pred)

                # Update lags recursively
                for lag in reversed(self.lags):
                    col = f"{f}_lag_{lag}"
                    if lag == 1:
                        last_row[col] = pred
                    else:
                        prev_lags = [l for l in self.lags if l < lag]
                        if prev_lags:
                            prev_col = f"{f}_lag_{max(prev_lags)}"
                            last_row[col] = last_row.get(prev_col, 0)

                # Update month/year
                if "month" in cols:
                    last_row["month"] = dt.month
                if "year" in cols:
                    last_row["year"] = dt.year

            self.predictions[f] = pd.Series(preds, index=future_index, name=f)

        # Compute total residuals (sum of all features)
        self.total_resid = pd.concat(self.predictions.values(), axis=1).sum(axis=1)
        self.total_resid.name = "total_resid"

        return self.predictions

    # -------------------------------------------------
    def plot_predictions(self, cumulative=False):
        if not self.predictions:
            raise ValueError("No predictions available. Run predict_future first.")

        plt.figure(figsize=(14, 7))
        for f, series in self.predictions.items():
            if series.sum() != 0:
                y = series.cumsum() if cumulative else series
                plt.plot(series.index, y, label=f"{f}", linewidth=2, linestyle='--')

        if self.total_resid is not None:
            y_total = self.total_resid.cumsum() if cumulative else self.total_resid
            plt.plot(self.total_resid.index, y_total, label="Total Residuals", linewidth=3, color='black')

        title = "Cumulative Predicted Residual Shares" if cumulative else "Predicted Residual Shares per Feature"
        plt.title(title)
        plt.xlabel("Month")
        plt.ylabel("Residual Value")
        plt.grid(True, linestyle="--", alpha=0.3)
        plt.legend()
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
aggregator = ResidualForecastAggregator(
    models, 
    specific_features, 
    feature_cols, 
    lags, 
    clip_std=1 # Adjust how strictly to clip extreme residuals
)

predictions = aggregator.predict_future(
    training_dfs,
    resid_series=resid_comp,  # historical residuals
    start_date="2026-01-31",
    periods=12
)

aggregator.plot_predictions(cumulative=False)
aggregator.plot_predictions(cumulative=True)

resid_forecast = aggregator.total_resid
print(f"Total residuals in millions: {resid_forecast.sum() / 1e6:.2f}")
```

```python
all_arima_order = []
total_of_resid_conterbution = 0
total_abs_of_resid_conterbution = 0
def forecast_all_ts(data_clean):
    global total_of_resid_conterbution
    global total_abs_of_resid_conterbution

    forecast_rows = []
    bad_ts_rows = []

    unique_combos = (
        data_clean
        .groupby(["huka", "expediture_type", fc_or_zroa_selected, "doc_type"])
        ["mimun"]
        .sum()
        .index
    )

    for combo in unique_combos:
        huka, exp_type, fc_val, doc_type = combo

        mask_clean = (
            (data_clean["huka"] == huka) &
            (data_clean["expediture_type"] == exp_type) &
            (data_clean[fc_or_zroa_selected] == fc_val) &
            (data_clean["doc_type"] == doc_type)
        )
        time_series = data_clean.loc[mask_clean].groupby("paymant_date")["mimun"].sum().resample("ME").sum().fillna(0)

        inspector = TimeSeriesInspector(time_series)
        clean_active_ts = inspector.get_clean_active_ts()
        clean_active_ts = fill_inactive_months(clean_active_ts)

        # -----------------------------
        # BAD TS
        # -----------------------------
        if clean_active_ts[clean_active_ts>0].count() < 12:
            bad_ts_rows.append({
                "huka": huka,
                "expediture_type": exp_type,
                fc_or_zroa_selected: fc_val,
                "doc_type": doc_type,
            })
        else:

            # -----------------------------
            # TREND + DIAGNOSTICS
            # -----------------------------
            trend, rolling_std = find_trend_and_rolling_std(clean_active_ts)
            slope, r2, elasticity, model_prediction = variance_trend_diagnostics(
                trend, rolling_std
            )

            decomposition_type, reason_of_decomposition = choose_decomposition(
                clean_active_ts, r2, elasticity
            )

            # FORCE ADDITIVE (your request)
            decomposition_type = "additive"

            decomposition = make_decompose(clean_active_ts, decomposition_type)
            trend_comp, seasonal_comp, resid_comp = extract_decomposition(decomposition)

            # -----------------------------
            # FORECAST HORIZON
            # -----------------------------
            last_date = trend_comp.index[-1]
            months_to_predict = (
                (YEAR_TO_PREDICT + 1 - last_date.year) * MONTHS_IN_YEAR
                - last_date.month
            )

            # -----------------------------
            # TREND FORECAST
            # -----------------------------
            options_of_models = ["regular_arima", "holt", "polynomial", "optimized_arima"]
            # choose optimized arima
            model_to_use = "optimized_arima"
            if model_to_use == "regular_arima":
                model = ARIMA(trend_comp, order = ARIMA_ORDER)
                results = model.fit()
                trend_forecast = results.forecast(steps=months_to_predict)
            elif model_to_use == "holt":
                trend_forecast = forecast_trend(trend_comp, periods=months_to_predict, method='holt')
            elif model_to_use == "polynomial":
                trend_forecast = forecast_trend(trend_comp, periods=months_to_predict, method='polynomial', poly_order=2)
            elif model_to_use == "optimized_arima":
                try:
                    model = pm.auto_arima(
                        y=trend_comp,
                        start_p=0, max_p=5,
                        start_q=0, max_q=5,
                        d=None,
                        seasonal=False,
                        stepwise=False,
                        trace=False,
                        error_action='ignore'
                    )
                    # Fitted pmdarima model
                    trend_forecast = model.predict(n_periods=months_to_predict)
                except:
                    # fallback to statsmodels ARIMA
                    default_order = (0,2,0)
                    sm_model = ARIMA(trend_comp, order=default_order)
                    results = sm_model.fit()
                    trend_forecast = results.forecast(steps=months_to_predict)
                all_arima_order.append(model.order)



            if trend_forecast.sum() < 0: 
                trend_forecast = get_non_negative_forecast(
                        trend_comp,
                        periods=12,
                        last_n_months=10,
                        initial_shrink=2,
                        max_shrink=10,
                    )
            if trend_forecast.sum() < 0:
                # if still negative, make the forcast zero
                trend_forecast = pd.Series(0, index=trend_forecast.index)
    
            
            # -----------------------------
            # SEASONAL FORECAST
            # -----------------------------
            last_season = seasonal_comp.iloc[-12:].values

            seasonal_forecast_values = np.tile(
                last_season,
                reps=(months_to_predict // MONTHS_IN_YEAR + 1)
            )[:months_to_predict]

            seasonal_forecast_index = pd.date_range(
                start=trend_forecast.index[0],
                periods=months_to_predict,
                freq="ME"
            )

            seasonal_forecast = pd.Series(
                seasonal_forecast_values,
                index=seasonal_forecast_index
            )
            # add to the forcast the month that already has a value
            # ----------------------------
            # FORCE FULL YEAR INSIDE FORECAST SERIES
            # ----------------------------

            # Get observed months of prediction year
            trend_existing_year = trend_comp[trend_comp.index.year == YEAR_TO_PREDICT]
            seasonal_existing_year = seasonal_comp[seasonal_comp.index.year == YEAR_TO_PREDICT]

            # Combine observed + forecast
            trend_forecast = pd.concat([trend_existing_year, trend_forecast]).sort_index()
            seasonal_forecast = pd.concat([seasonal_existing_year, seasonal_forecast]).sort_index()

            # Keep only prediction year
            trend_forecast = trend_forecast[trend_forecast.index.year == YEAR_TO_PREDICT]
            seasonal_forecast = seasonal_forecast[seasonal_forecast.index.year == YEAR_TO_PREDICT]

            # -----------------------------
            # resid forecast
            # -----------------------------
            # on smah
            # base forecast (NO resid)
            forecast_monthly_no_resid = trend_forecast + seasonal_forecast

            if (doc_type == "RV") & (resid_comp[resid_comp.values != 0].count() > 24):
                #specific_features = ["inviting_partner", "sales_group", "sales_office", "sales_organistion"]
                specific_features = ["inviting_partner"]

                lags = (1, 3, 6, 12)

                # Dictionaries to hold models, feature columns, and training data. 
                models = {}
                feature_cols = {}
                training_dfs = {}

                models, feature_cols, training_dfs = train_models(specific_features, lags, huka, exp_type, fc_val, doc_type, resid_comp, clean_active_ts)

                aggregator = ResidualForecastAggregator(models, specific_features, feature_cols, lags)


                # Predict 12 months
                aggregator.predict_future(training_dfs,resid_comp, start_date="2026-01-31", periods=12)

                resid_forecast = aggregator.total_resid

                total_of_resid_conterbution += resid_forecast.sum()

                total_abs_of_resid_conterbution += resid_forecast.abs().sum()

                forecast_monthly = forecast_monthly_no_resid + resid_forecast
                
            # not on smah
            else:
                forecast_monthly = forecast_monthly_no_resid

            # MONTHLY FORECAST (ADDITIVE)
            # -----------------------------
            forecast_monthly = forecast_monthly[
                forecast_monthly.index.year == YEAR_TO_PREDICT
            ]

            if forecast_monthly.sum() < 0:
                # if still negative, make the forcast zero
                forecast_monthly = pd.Series(0, index=forecast_monthly.index)

            # -----------------------------
            # STORE ROW
            # -----------------------------
            row = {
                "huka": huka,
                "expediture_type": exp_type,
                fc_or_zroa_selected: fc_val,
                "doc_type": doc_type,
            }

            # with resid (existing behavior)
            for dt, val in forecast_monthly.items():
                row[dt.strftime("%Y-%m")] = val

            # without resid (NEW)
            for dt, val in forecast_monthly_no_resid.items():
                row[f"base_{dt.strftime('%Y-%m')}"] = val

            forecast_rows.append(row)

    # -----------------------------
    # BUILD FINAL DATAFRAMES
    # -----------------------------
    forecast_ts = (
        pd.DataFrame(forecast_rows)
        .set_index(["huka", "expediture_type", fc_or_zroa_selected, "doc_type"])
        .sort_index()
    )

    bad_ts = (
        pd.DataFrame(bad_ts_rows)
        .set_index(["huka", "expediture_type", fc_or_zroa_selected, "doc_type"])
        .sort_index()
    )
    # add to bad rows forcasted month that will be mean of of the series in years = [2023, 2024, 2025]
    for idx, row in bad_ts.iterrows():
        huka = idx[0]
        exp_type = idx[1]
        fc_val = idx[2]
        doc_type = idx[3]

        mask_original = (
            (data_clean["huka"] == huka) &
            (data_clean["expediture_type"] == exp_type) &
            (data_clean[fc_or_zroa_selected] == fc_val) &
            (data_clean["doc_type"] == doc_type)
        )
        time_series_original = data_clean.loc[mask_original].groupby("paymant_date")["mimun"].sum().resample("YE").sum().fillna(0)
        mean_value = time_series_original[
            (time_series_original.index.year >= YEAR_TO_PREDICT - 2) &
            (time_series_original.index.year <= YEAR_TO_PREDICT)
        ].mean()

        for month in range(1, 13):
            month_str = f"{YEAR_TO_PREDICT}-{month:02d}"
            bad_ts.at[idx, month_str] = mean_value / 12

    return forecast_ts, bad_ts.fillna(0)


original_data_daily = (
    tashlomim_data_specific_coin
    .groupby(FULL_KEYS, as_index=False)
    .agg(mimun=('mimun', 'sum'))
    .sort_values('paymant_date')
)

spike_detector = SpikeDetection(original_data_daily.copy())
all_data_daily, flagged_as_spikes, data_clean = spike_detector.find_spikes()

data_clean = data_clean[ FULL_KEYS + ['mimun'] ]

forecasted_df, bad_ts = forecast_all_ts(data_clean)

# add bad ts to forecasted df
all_forecasted_df = pd.concat([forecasted_df, bad_ts], axis=0)
# yearly WITH resid
all_forecasted_df["forcast_yearly"] = (
    all_forecasted_df
    .filter(regex=f"^{YEAR_TO_PREDICT}-")
    .sum(axis=1)
)

# yearly WITHOUT resid (trend + seasonality only)
all_forecasted_df["forcast_yearly_no_resid"] = (
    all_forecasted_df
    .filter(regex=f"^base_{YEAR_TO_PREDICT}-")
    .sum(axis=1)
)# sum in total forecasted value
total_forecasted_value = all_forecasted_df["forcast_yearly"].sum()
just_yearly_forcast = all_forecasted_df[
    ["forcast_yearly", "forcast_yearly_no_resid"]
]

# add actual 2025 values to just_yearly_forcast
# years to add
years = [2023, 2024, 2025]

for year in years:
    values_for_year = (
        tashlomim_data_orginal_before_removing_biden[
            tashlomim_data_orginal_before_removing_biden['paymant_date'].dt.year == year
        ]
        .groupby(KEYS, as_index=False)
        .agg(mimun=('mimun', 'sum'))
        .set_index(['huka', 'expediture_type', fc_or_zroa_selected, 'doc_type'])
        .rename(columns={'mimun': f'actual_{year}'})
    )

    just_yearly_forcast = just_yearly_forcast.merge(
        values_for_year[[f'actual_{year}']],
        left_index=True,
        right_index=True,
        how='outer'
    )
just_yearly_forcast = just_yearly_forcast.fillna(0)
just_yearly_forcast = just_yearly_forcast[~((just_yearly_forcast["actual_2023"] == 0) & (just_yearly_forcast["actual_2024"] == 0) & (just_yearly_forcast["actual_2025"] == 0))]

# temprorary fix
# if netetive in forcast_yearly sum change it to what was last year
mask = just_yearly_forcast["forcast_yearly"] < 0
just_yearly_forcast.loc[mask, "forcast_yearly"] = just_yearly_forcast.loc[mask, "actual_2025"]

display(Markdown(f"## Total forecasted value for year {YEAR_TO_PREDICT}: {just_yearly_forcast['forcast_yearly'].sum():,.2f}"))
# rename columns
just_yearly_forcast.rename(columns={'forcast_yearly': 'total_forecasted_value'}, inplace=True)
just_yearly_forcast.rename(columns={'forcast_yearly_no_resid': 'total_forecasted_value_no_resid'}, inplace=True)
# save result
just_yearly_forcast.to_csv(f"yearly_forecast_vs_actual_{COIN_TO_INSPECT}.csv")
```

```python
def plot_by_feature(by_this_feature_agg):
    # -----------------------------
    # Aggregate forecast
    # -----------------------------
    prediction_yearly_forcast_agg = (
        just_yearly_forcast
        .reset_index()
        .groupby(by_this_feature_agg)
        .agg({'total_forecasted_value': 'sum'})
        .sort_values('total_forecasted_value', ascending=False)
    )

    # -----------------------------
    # Aggregate historical data
    # -----------------------------
    tashlomim_data_specific_coin_agg = (
        tashlomim_data_specific_coin
        .groupby([by_this_feature_agg, 'payment_year'])
        .agg({'mimun': 'sum'})
        .reset_index()
    )

    # -----------------------------
    # Append forecast year
    # -----------------------------
    for feature_value, row in prediction_yearly_forcast_agg.iterrows():
        tashlomim_data_specific_coin_agg = pd.concat(
            [
                tashlomim_data_specific_coin_agg,
                pd.DataFrame({
                    by_this_feature_agg: [feature_value],
                    'payment_year': [YEAR_TO_PREDICT],
                    'mimun': [row['total_forecasted_value']]
                })
            ],
            ignore_index=True
        )

    # -----------------------------
    # Plot top 10 features
    # -----------------------------
    top_n = 10
    features_to_plot = prediction_yearly_forcast_agg.head(top_n).index

    fig, axes = plt.subplots(
        nrows=5,
        ncols=2,
        figsize=(16, 20),
        squeeze=False
    )

    for i, feature_value in enumerate(features_to_plot):
        ts_feature = (
            tashlomim_data_specific_coin_agg[
                tashlomim_data_specific_coin_agg[by_this_feature_agg] == feature_value
            ]
            .groupby('payment_year')
            .agg({'mimun': 'sum'})
        )

        ax = axes[i // 2, i % 2]

        # -----------------------------
        # Split historical vs forecast
        # -----------------------------
        hist = ts_feature[ts_feature.index < YEAR_TO_PREDICT]
        forecast = ts_feature[ts_feature.index >= YEAR_TO_PREDICT]

        # Historical
        ax.plot(
            hist.index,
            hist['mimun'],
            marker='o',
            linewidth=2,
            label='Actual'
        )

        # Forecast (different style & color)
        ax.plot(
            forecast.index,
            forecast['mimun'],
            marker='o',
            linestyle='--',
            linewidth=3,
            color='crimson',
            label='Forecast'
        )

        ax.set_title(f'{by_this_feature_agg} = {feature_value}', fontsize=12, weight='bold')
        ax.set_xlabel('Year')
        ax.set_ylabel('Total Mimun')
        ax.grid(alpha=0.3)
        ax.set_xticks(ts_feature.index)
        ax.legend()

    plt.suptitle(
        f'Yearly Mimun by {by_this_feature_agg} (Forecast Highlighted)',
        fontsize=16,
        weight='bold'
    )
    # delete unused subplots
    for j in range(i + 1, 10):
        fig.delaxes(axes[j // 2, j % 2])
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()
feature_widget = widgets.Dropdown(
    options=['fc3', 'huka', 'expediture_type', 'doc_type'],
    value='huka',
    description='Group by:',
    style={'description_width': 'initial'}
)

widgets.interact(
    plot_by_feature,
    by_this_feature_agg=feature_widget
)
```
