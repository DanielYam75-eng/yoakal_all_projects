```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
invoices = pd.read_csv("NN-data-5-15-1.csv")
orders = pd.read_csv("orders coin 1 article 15.csv")
```

```python
invoices["doc type"] = "invoices"
orders["doc type"] = "orders"
```

```python
VAL = "volume"
YEAR = "year"
MONTH = "month"
GROUP = "doc"
```

```python
orders = orders[orders[YEAR] >= invoices[YEAR].min()]
```

```python
orders = orders.groupby(orders.columns.difference([VAL]).to_list()).sum().reset_index()
```

```python
combined = pd.concat([invoices, orders])
```

```python
combined.tail()
```

```python
combined = combined.set_index(pd.to_datetime(combined[YEAR].astype(str) + "-" + combined[MONTH].astype(str))).sort_index()
```

```python
combined.head()
```

```python
example = combined[combined[GROUP].isin(combined[GROUP].unique()[:8])]

for group, data in example.groupby(GROUP):

    data.resample("YE")[VAL].sum().plot(label = group)

plt.legend()
```

```python
combined.reset_index(drop = True).to_csv("combined_data-5-15-1.csv", index = False)
```
